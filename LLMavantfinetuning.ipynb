Le modèle distilbert-base-uncased est un modèle de traitement du langage naturel (NLP) qui fait partie de la famille des modèles BERT (Bidirectional Encoder Representations from Transformers).

On a chargé le modèle et le tokenizer, appliqué le pruning et la quantization, puis créé une fonction pour générer des réponses aux questions.

#### Pruning : Le pruning, ou élagage, est une technique qui réduit la taille d'un modèle en supprimant des poids ou des neurones jugés peu importants, ce qui améliore son efficacité et réduit l'utilisation de la mémoire.


#### Quantization:La quantization consiste à convertir les poids et les activations d'un modèle de formats à virgule flottante en formats entiers, ce qui réduit leur précision. Cette technique permet de diminuer la taille du modèle et d'accélérer les temps de calcul tout en préservant une précision acceptable.

````python
def load_model_and_tokenizer(model_id):
    """
    Load the question-answering model and tokenizer from the given model ID.

    :param model_id: Model ID to load (e.g., 'distilbert-base-uncased').
    :return: Loaded model and tokenizer.
    """
    model = AutoModelForQuestionAnswering.from_pretrained(model_id)
    tokenizer = AutoTokenizer.from_pretrained(model_id)
    return model, tokenizer

def apply_pruning(model, amount=0.2):
    """
    Apply pruning to the linear layers of the model.

    :param model: The model to prune.
    :param amount: The fraction of weights to prune (default 20%).
    """
    for name, module in model.named_modules():
        if isinstance(module, torch.nn.Linear):
            prune.l1_unstructured(module, name='weight', amount=amount)

def quantize_model(model):
    """
    Apply dynamic quantization to the model to reduce weight size.

    :param model: The model to quantize.
    :return: Quantized model.
    """
    model.eval()
    quantized_model = torch.quantization.quantize_dynamic(model, {torch.nn.Linear}, dtype=torch.qint8)
    return quantized_model

def get_answer(question, context, model, tokenizer, device):
    """
    Generate an answer to a given question using the question-answering model.

    :param question: Question to ask.
    :param context: Context to search for the answer.
    :param model: Question-answering model.
    :param tokenizer: Tokenizer for encoding questions and context.
    :param device: Device (CPU or GPU) to run the model.
    :return: Generated answer.
    """
    inputs = tokenizer.encode_plus(question, context, return_tensors='pt')
    input_ids = inputs['input_ids'].to(device)
    attention_mask = inputs['attention_mask'].to(device)

    with torch.no_grad():
        outputs = model(input_ids, attention_mask=attention_mask)

    start_scores, end_scores = outputs.start_logits, outputs.end_logits
    start_index = torch.argmax(start_scores)
    end_index = torch.argmax(end_scores)

    answer_tokens = input_ids[0][start_index:end_index + 1]
    answer = tokenizer.decode(answer_tokens, skip_special_tokens=True)

    return answer

````

````python
model_id = "distilbert-base-uncased"
model, tokenizer = load_model_and_tokenizer(model_id)

quantized_model = quantize_model(model)
device = torch.device("cpu")

apply_pruning(quantized_model)

train_dataset = dataset['train']

for i in range(10):
    question = train_dataset[i]['question']
    context = train_dataset[i]['context']
    answer = get_answer(question, context, quantized_model, tokenizer, device)
    print(f"Question {i + 1}: {question}\nAnswer: {answer}\n")
````

````python
from rouge_score import rouge_scorer
from time import time

def test_model(tokenizer, pipeline, prompt_to_test, ground_truths):
    """
    Perform a query and calculate ROUGE scores.

    Args:
        tokenizer: the tokenizer
        pipeline: the pipeline
        prompt_to_test: the prompt
        ground_truths: list of expected outputs
    Returns:
        None
    """
    time_1 = time()
    sequences = pipeline(
        prompt_to_test,
        do_sample=True,
        top_k=10,
        num_return_sequences=1,
        eos_token_id=tokenizer.eos_token_id,
        max_length=200,
    )
    time_2 = time()
    print(f"Test inference: {round(time_2 - time_1, 3)} sec.")

    model_outputs = [seq['generated_text'] for seq in sequences]

    for output in model_outputs:
        print(f"Result: {output}")

    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

    for output in model_outputs:
        for ground_truth in ground_truths:
            scores = scorer.score(ground_truth, output)
            print(f"ROUGE Scores for output'{output}':")
            print(f"  ROUGE-1: {scores['rouge1'].fmeasure:.4f}")
            print(f"  ROUGE-2: {scores['rouge2'].fmeasure:.4f}")
            print(f"  ROUGE-L: {scores['rougeL'].fmeasure:.4f}")

````

## *Split* the data into chunks¶



<div style="background-color: #87CEEB; padding: 10px; border-radius: 5px;">
=> Nous avons pris un texte relativement long, issu de SQuAD v2.0 (une base de données utilisée pour l'entraînement de modèles de questions-réponses), et l'avons divisé en plusieurs "chunks" de taille maximale définie (ici 512 tokens). Le but de cette opération est de rendre le traitement plus efficace pour les modèles de langage qui ont des limitations sur la longueur du texte pouvant être analysé simultanément (comme GPT, BERT, etc.).

````python
max_chunk_size = 512
def chunk_text(text, max_chunk_size):
    """Split the input text into chunks."""
    words = text.split()
    chunks = []
    current_chunk = []
    current_length = 0

    for word in words:
        word_length = len(word) + 1
        if current_length + word_length > max_chunk_size:
            chunks.append(" ".join(current_chunk))
            current_chunk = [word]
            current_length = word_length
        else:
            current_chunk.append(word)
            current_length += word_length

    if current_chunk:
        chunks.append(" ".join(current_chunk))

    return chunks

def chunk_examples(example):
    context_chunks = chunk_text(example['context'], max_chunk_size)
    return {'question': example['question'], 'context_chunks': context_chunks}

chunked_dataset = dataset['train'].map(chunk_examples)

print("Example of chunked data:")
for i in range(3):
    print(chunked_dataset[i])
````

**Conclusion**

La stratégie de découpage en chunks présente plusieurs avantages :

*Réduction de la complexité pour les modèles NLP :*

Les grands modèles de traitement du langage, comme GPT ou BERT, ont une limitation sur la longueur du texte qu'ils peuvent traiter en une seule fois. En divisant le texte en fragments, on permet au modèle de mieux comprendre le contexte sans être submergé.

*Amélioration de la précision des réponses :*

 En travaillant sur des morceaux plus petits et plus ciblés, il devient plus facile d'isoler les informations pertinentes pour répondre à une question spécifique.

*Optimisation de la mémoire :*


Lors de l'entraînement ou de l'inférence avec de grandes quantités de données, travailler avec des fragments permet d'utiliser la mémoire plus efficacement.
Ce découpage en fragments rend donc les systèmes de génération de réponses ou de récupération de données plus performants et précis dans des applications comme les systèmes de questions-réponses, la recherche d'information ou les systèmes de génération de texte augmentés par la récupération (RAG).

````python
import torch
from langchain import PromptTemplate, LLMChain
from langchain.llms import HuggingFacePipeline
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

huggingface_llm = HuggingFacePipeline(pipeline=query_pipeline)

prompt_template = PromptTemplate(
    template="Question: {question}\nContext: {context}\nAnswer:",
    input_variables=["question", "context"]
)

llm_chain = LLMChain(llm=huggingface_llm, prompt=prompt_template)

example_question = "Do you think my credit card is stolen?"
example_context = "Your credit card details should be kept secure. If you notice any unauthorized transactions, report them immediately."

try:
    response = llm_chain.run({"question": example_question, "context": example_context})

    print("Generated Response:", response)
except Exception as e:
    print(f"An error occurred: {e}")
````

````python
def apply_langchain_to_chunks(example):
    responses = []
    for context_chunk in example['context_chunks']:
        response = query_pipeline(context_chunk, do_sample=True, top_k=10, max_new_tokens=50)
        responses.append(response[0]['generated_text'])
    return {'question': example['question'], 'responses': responses}
````

# 1. Différence entre génération Hugging Face et LangChain

La structure imposée par LangChain a pour effet de guider la réponse, ce qui réduit la longueur ou la spontanéité de la réponse au profit de quelque chose de plus précis et direct.

Pourquoi ?

Cela est dû à la façon dont LangChain fonctionne. Le modèle est construit autour de "templates" qui encadrent la génération et guident la réponse, tandis qu'avec un modèle Hugging Face direct, la génération est moins encadrée, ce qui peut produire des réponses plus créatives mais aussi plus longues et potentiellement moins pertinentes dans certains contextes.



# 2. Génération et structure dans LangChain



**La structure qu'on a imposé à LangChain, en utilisant un PromptTemplate tel que :**

- Question: {question}

- Context: {context}

- Answer:
a un effet sur la manière dont le modèle génère ses réponses. Dans ce cas, le modèle suit strictement la consigne du prompt et répond plus directement à la question en se basant sur le contexte fourni.

- Conclusion
Les résultats obtenus montrent que l'application de LangChain pour guider les réponses avec un template structuré peut produire des réponses plus concises, précises, et orientées sur la question. En parallèle, l'utilisation de la génération en "chunks" aide à mieux gérer de longs contextes en découpant les informations et en les traitant morceau par morceau, garantissant ainsi des réponses plus précises pour chaque partie du texte.

````python
model_id = 'meta-llama/Llama-2-7b-chat-hf'

def prepare_model_and_pipeline(model_id):
    device_map = {
        "": "cuda:0",
        "transformer.h.0": "cuda:0",
        "transformer.h.1": "cpu",
        "transformer.h.2": "cpu",
    }

    model = AutoModelForCausalLM.from_pretrained(
        model_id,
        load_in_8bit=True,
        device_map=device_map,
        llm_int8_enable_fp32_cpu_offload=True
    )

    tokenizer = AutoTokenizer.from_pretrained(model_id)

    text_generation_pipeline = pipeline("text-generation", model=model, tokenizer=tokenizer)

    print("Model and pipeline loaded successfully.")
    return text_generation_pipeline

try:
    text_gen_pipeline = prepare_model_and_pipeline(model_id)
except Exception as e:
    print(f"Error loading model: {e}")
    text_gen_pipeline = None

def generate_text(prompt, max_length=50):
    if text_gen_pipeline is None:
        return "Pipeline not set up. Please check the model loading process."
    try:
        generated_outputs = text_gen_pipeline(prompt, max_length=max_length, num_return_sequences=1)
        return generated_outputs[0]['generated_text']
    except Exception as e:
        return f"Error generating text: {e}"


````
